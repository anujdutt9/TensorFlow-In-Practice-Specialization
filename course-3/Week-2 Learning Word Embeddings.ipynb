```python
# Import Dependencies
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Check TF version
tf.__version__
```

```python
# Load IMDB Reviews Dataset using TFDS
imdb, info = tfds.load(name="imdb_reviews", with_info=True, as_supervised=True)
```

```python
# Dataset Info
info
```

```python
# Get Training and Test Dataset
train_data, test_data = imdb['train'], imdb['test']
```

```python
len(train_data)
```

```python
len(test_data)
```

```python
training_sentences = []
training_labels = []

test_sentences = []
test_labels = []
```

```python
# Get Training Dataset
for s, l in train_data:
    training_sentences.append(str(s.numpy()))
    training_labels.append(l.numpy())
```

```python
# Get Training Dataset
for s, l in test_data:
    test_sentences.append(str(s.numpy()))
    test_labels.append(l.numpy())
```

```python
# Convert Labels to Numpy Array
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(test_labels)
```

```python
training_labels_final.shape
```

```python
testing_labels_final.shape
```

```python
# Define Hyperparameters

# Vocabulary Size
vocab_size = 10000

# Embeddings Dimension
embedding_dim = 16

# Sentence Embedding Max Length
max_length = 120

# Truncation Type for Word Embedding
trunc_type='post'

# OOV: Out of Vocabulary, for text in input not in vocabulary
oov_tok = "<OOV>"
```

```python
# Tokenize the Sentences in Training Dataset
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)

# Get Word to Index Mapping for Tokens
word_index = tokenizer.word_index

# Generate Text to Word Embedding Sequence Mapping
sequences = tokenizer.texts_to_sequences(training_sentences)

# Create Padded Word Embedding Sequence
padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type)

# Tokenize and Pad Test Dataset
testing_sequences = tokenizer.texts_to_sequences(test_sentences)
testing_padded = pad_sequences(testing_sequences,maxlen=max_length)
```

```python
# Index to Word Mapping
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
reverse_word_index
```

```python
# Function to decode Padded Reviews i.e. Index to Word Mapping
# '?' denotes the padding in front, you can replace that with '0' as well
def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])
```

```python
# Decoding Sample Padded Review from Training Dataset
print(decode_review(padded[3]))
print(training_sentences[3])
```

```python
# Model Definition with Flatten Layer
model1 = tf.keras.Sequential([
    # Embedding Layer to learn word vector embeddings/relationships
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model1.summary()
```

```python
# Model Definition with Global Average Pooling 1D Layer instead of Flatten Layer
model2 = tf.keras.Sequential([
    # Embedding Layer to learn word vector embeddings/relationships
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model2.summary()
```

```python
# Compile both the Models
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Number of Epochs to train the Model for
num_epochs = 10

# Train the Model-1
# Training Data: Padded Word Vocabulary Values, Labels as Numpy Array
# Validation Data: Padded Word Vocab Values, Labels as Numpy Array
model1.fit(padded, training_labels_final, 
          epochs=num_epochs, 
          validation_data=(testing_padded, testing_labels_final))
```

```python
# Train the Model-1
# Training Data: Padded Word Vocabulary Values, Labels as Numpy Array
# Validation Data: Padded Word Vocab Values, Labels as Numpy Array
model2.fit(padded, training_labels_final, 
          epochs=num_epochs, 
          validation_data=(testing_padded, testing_labels_final))
```

As we can see, the model with GlobalAveragePooling1D layer trains way faster [2ms per epoch] than the one with the Flatten Layer [1 sec per epoch].

```python
# Get Trained Embeddings from the Model
# The Embeddings are learnt as part of the Model Training Process
embedding = model1.layers[0]
weights = embedding.get_weights()[0]

# shape: (vocab_size, embedding_dim)
print(weights.shape)
```

## Visualize the Trained Embeddings

```python
# Save Trained Embeddings
import io

out_v = io.open('vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('meta.tsv', 'w', encoding='utf-8')

# For all words in the vocabulary
for word_num in range(1, vocab_size):
    # Get the words for index
    word = reverse_word_index[word_num]
    # Pass in the word through the Embedding Layer Only
    embeddings = weights[word_num]
    print('Word: {0}\t Embeddings: {1}\n'.format(word, embeddings))
    out_m.write(word + "\n")
    out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")

out_v.close()
out_m.close()
```

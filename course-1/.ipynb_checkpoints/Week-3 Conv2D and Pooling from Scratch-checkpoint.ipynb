```python
import cv2
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

```python
i = misc.ascent()
```

```python
# Show the Image
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(i)
plt.show()
```

```python
# Copy the Image as numpy array and get it's size
i_transformed = np.copy(i)
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]
```

```python
size_x, size_y
```

```python
# Define Conv2D Filters
filter1 = [ [0, 1, 0], [1, -4, 1], [0, 1, 0]]
filter2 = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]
filter3 = [ [-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
```

```python
# If all the digits in the filter don't add up to 0 or 1, you 
# should probably do a weight to get it to do so
# so, for example, if your weights are 1,1,1 1,2,1 1,1,1
# They add up to 10, so you would set a weight of .1 if you want to normalize them
weight  = 1
```

```python
# Function to perform Conv2D operation
def Conv2D(size_x=None, size_y=None, convFilter=None):
    # We start with index of 1 and end at size - 1
    for x in range(1,size_x-1):
        for y in range(1,size_y-1):
            convolution = 0.0
            convolution = convolution + (i[x - 1, y-1] * convFilter[0][0])
            convolution = convolution + (i[x, y-1] * convFilter[0][1])
            convolution = convolution + (i[x + 1, y-1] * convFilter[0][2])
            convolution = convolution + (i[x-1, y] * convFilter[1][0])
            convolution = convolution + (i[x, y] * convFilter[1][1])
            convolution = convolution + (i[x+1, y] * convFilter[1][2])
            convolution = convolution + (i[x-1, y+1] * convFilter[2][0])
            convolution = convolution + (i[x, y+1] * convFilter[2][1])
            convolution = convolution + (i[x+1, y+1] * convFilter[2][2])
            convolution = convolution * weight
            
            if(convolution<0):
                convolution=0
            
            if(convolution>255):
                convolution=255
            
            i_transformed[x, y] = convolution
    return i_transformed
```

### Image -> Conv2D -> Output

```python
# Plot the image after applying filter1 to it
i_transformed = Conv2D(size_x, size_y, filter1)
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show()  
```

```python
# Plot the image after applying filter2 to it
i_transformed = Conv2D(size_x, size_y, filter2)
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show() 
```

```python
# Plot the image after applying filter3 to it
i_transformed = Conv2D(size_x, size_y, filter3)
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show() 
```

```python
# MaxPool2D operation
def MaxPool2D(size_x=None, size_y=None, i_transformed=None):
    new_x = int(size_x/2)
    new_y = int(size_y/2)
    newImage = np.zeros((new_x, new_y))
    for x in range(0, size_x, 2):
        for y in range(0, size_y, 2):
            pixels = []
            pixels.append(i_transformed[x, y])
            pixels.append(i_transformed[x+1, y])
            pixels.append(i_transformed[x, y+1])
            pixels.append(i_transformed[x+1, y+1])
            newImage[int(x/2),int(y/2)] = max(pixels)
    return newImage
```

### Image - > Conv2D -> MaxPool2D -> Output

```python
# Plot the image after applying Conv2D + MaxPool2D to it
i_transformed = Conv2D(size_x, size_y, filter1)
newImage = MaxPool2D(size_x, size_y, i_transformed)
plt.gray()
plt.grid(False)
plt.imshow(newImage)
plt.show()  
```

```python
# Plot the image after applying Conv2D + MaxPool2D to it
i_transformed = Conv2D(size_x, size_y, filter2)
newImage = MaxPool2D(size_x, size_y, i_transformed)
plt.gray()
plt.grid(False)
plt.imshow(newImage)
plt.show()  
```

```python
# Plot the image after applying Conv2D + MaxPool2D to it
i_transformed = Conv2D(size_x, size_y, filter3)
newImage = MaxPool2D(size_x, size_y, i_transformed)
plt.gray()
plt.grid(False)
plt.imshow(newImage)
plt.show()  
```
